```python
from aocd import get_data
data = get_data(day=2, year=2022).split('\n')
```

```python

opponent_score_dict = {'A':1, 'B':2, 'C':3}
my_score_dict = {'X':1, 'Y':2, 'Z':3}

opponent_shape_list = ['A','C','B']
my_shape_list = ['X','Z','Y']

translation_dict = {0: 'Rock', 1:'Scissor', 2: 'Paper'}

tmp_lst = []
for i in data:
    tmp = i.split(' ')
    tmp_lst.append(tmp)
```

```python
score_this_round = 0
total_score = 0

for i in tmp_lst:
    first = i[0]
    second = i[1]
    opponent_score= opponent_score_dict[first]
    me_score= my_score_dict[second]
    opponent_shape = opponent_shape_list.index(i[0])
    
    me_shape = my_shape_list.index(i[1])
    #print(f'me: {translation_dict[me_shape]}, opponent: {translation_dict[opponent_shape]}\nme shape: {me_shape}, opponent shape:{opponent_shape}')
    if me_shape-opponent_shape== -1 or me_shape - opponent_shape ==2 :
        print("Me wins!")
        score_this_round = 6+me_score
    elif me_shape == opponent_shape:
        print("A draw!")
        score_this_round = 3+me_score
    else:
        print("Me lozes :( ") 
        score_this_round = me_score
    #print(score_this_round)
    total_score+= score_this_round


print(total_score)   
```
